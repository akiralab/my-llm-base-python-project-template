```python
# Example: synthetic data creation
SEED = 42
rng = np.random.default_rng(SEED)

df = pd.DataFrame(
    {
        'x': rng.normal(loc=0.0, scale=1.0, size=200),
        'y': rng.normal(loc=1.0, scale=2.0, size=200),
        'group': rng.choice(['A', 'B'], size=200),
    }
)

df.head()
```

```python
# Example: quick visualization and summary
fig, ax = plt.subplots(figsize=(7, 4))
sns.scatterplot(data=df, x='x', y='y', hue='group', alpha=0.7, ax=ax)
ax.set_title('Template Scatter Plot')
plt.show()

summary = df.groupby('group').agg(['mean', 'std'])
summary
```
